#### Using the Kalman Filter to Find Exact Finite-Sample Forecasts for an MA(1) Process
pp. 382 - *Time Series Analysis* by James Hamilton

Consider the state-space representation for the MA(1) process:<br>
<br>
**State Equation** (r = 2): <br>
<br>
$~~~~~~~~~~~~~~~\begin{bmatrix}\varepsilon_{t+1} \\ \varepsilon_t \end{bmatrix} = \begin{bmatrix}0 & 0\\1 & 0\end{bmatrix}\begin{bmatrix}\varepsilon_{t} \\ \varepsilon_{t-1} \end{bmatrix} + \begin{bmatrix}\varepsilon_{t+1} \\ 0 \end{bmatrix}$ <br>
<br>
**Observation Equation** (n = 1): <br>
<br>
$~~~~~~~~~~~~~~~y_t = \mu + \begin{bmatrix}1 & \theta \end{bmatrix}\begin{bmatrix}\varepsilon_{t} \\ \varepsilon_{t-1} \end{bmatrix}$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{\xi}_t = \begin{bmatrix}\varepsilon_{t} \\ \varepsilon_{t-1} \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{F} = \begin{bmatrix} 0 & 0 \\ 1 & 0\end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{v}_{t + 1} = \begin{bmatrix} \varepsilon_{t + 1} \\ 0\end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{Q} = \begin{bmatrix} \sigma^2 & 0 \\ 0 & 0\end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{y}_t = y_t$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{A'} = \mu$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{x}_t = 1$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{H'} = \begin{bmatrix} 1 & \theta\end{bmatrix}$
<br>
$~~~~~~~~~~~~~~~\mathbf{w}_t = 0$
<br>
$~~~~~~~~~~~~~~~\mathbf{R} = 0$<br>

If all of the eigenvalues of $\mathbf{F}$ are in the unit circles, then good starting values are:<br>
<br>
$~~~~~~~~~~~~\mathbf{\xi}_{1~|~0} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~\mathbf{P}_{1~|~0} = \begin{bmatrix} \sigma^2 & 0 \\ 0 & \sigma^2 \end{bmatrix}$

The period 1 forecast is:<br>
<br>
$~~~~~~~~~~~\hat{\mathbf{y}}_{t+1|t} = \mathbf{A'x}_{t+1} + \mathbf{H'\hat{\xi}}_{t+1|t}$<br>
<br>
$~~~~~~~~~~~\hat{\mathbf{y}}_{t+1|t} = \mu \times 1 + \begin{bmatrix} 1 & \theta\end{bmatrix}\begin{bmatrix} 0 \\ 0 \end{bmatrix} = \mu$

The mean squared error (MSE) is given by:<br>
<br>
$~~~~~~~~~~~~E[(\mathbf{\hat{y}}_{t+1} - \mathbf{\hat{y}}_{t+1|t})(\mathbf{\hat{y}}_{t+1} - \mathbf{\hat{y}}_{t+1|t})'] = \mathbf{H'P}_{t+1|t}\mathbf{H} + \mathbf{R}$<br>
<br>
$~~~~~~~~~~~~= \begin{bmatrix} 1 & \theta\end{bmatrix}\begin{bmatrix} \sigma^2 & 0 \\ 0 & \sigma^2 \end{bmatrix}\begin{bmatrix} 1 \\ \theta\end{bmatrix} + 0$ <br>
<br>
$~~~~~~~~~~~~= \sigma^2(1 + \theta^2)$

The above values are just the unconditional mean and variance of y.<br>

We now iterate on:<br>
<br>
$\mathbf{\hat{\xi}}_{t+1|t} = \mathbf{F\hat{\xi}}_{t|t-1} + \mathbf{FP}_{t|t-1}\mathbf{H}(\mathbf{H'P}_{t|t-1}\mathbf{H} + \mathbf{R})^{-1}(\mathbf{y}_t - \mathbf{A'x}_t - \mathbf{H'\hat{\xi}}_{t|t-1})$<br>
<br>
and plug that into:<br>
<br>
$\mathbf{\hat{y}}_{t+1|t} = \mathbf{A'x}_{t+1} + \mathbf{H'\hat{\xi}}_{t+1|t}$

For the first iteration of this problem, <br>
<br>
$\mathbf{\hat{\xi}}_{2|1} = \begin{bmatrix} 0 & 0 \\ 1 & 0\end{bmatrix} \begin{bmatrix} 0 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 & 0 \\ 1 & 0\end{bmatrix}\begin{bmatrix} \sigma^2 & 0 \\ 0 & \sigma^2 \end{bmatrix}\begin{bmatrix} 1 \\ \theta\end{bmatrix} \left( \begin{bmatrix} 1 & \theta\end{bmatrix}\begin{bmatrix} \sigma^2 & 0 \\ 0 & \sigma^2 \end{bmatrix}\begin{bmatrix} 1 \\ \theta\end{bmatrix} + O \right)^{-1}\left(y_t - \mu\times 1 - \begin{bmatrix} 1 & \theta\end{bmatrix}\begin{bmatrix} 0 \\ 0 \end{bmatrix}\right)$<br>
<br>
$= 0 + \begin{bmatrix} 0 & 0 \\ \sigma^2 & 0\end{bmatrix}\begin{bmatrix} 1 \\ \theta \end{bmatrix}\left( \begin{bmatrix} \sigma^2 & \theta\sigma^2\end{bmatrix}\begin{bmatrix} 1 \\ \theta\end{bmatrix} + O \right)^{-1}\left(y_t - \mu - 0\right)$<br>
<br>
$= \begin{bmatrix} 0 \\ \sigma^2 \end{bmatrix}\left( \sigma^2(1 + \theta^2)\right)^{-1}\left(y_t - \mu\right)$<br>
<br>
$= \begin{bmatrix} 0 \\ \dfrac{\left(y_t - \mu\right)}{1 + \theta^2} \end{bmatrix} = \begin{bmatrix} \mathbf{\hat{\varepsilon}}_{2|1} \\ \mathbf{\hat{\varepsilon}}_{1|1}\end{bmatrix}$


Since the first row of $\mathbf{F}$  consists entirely of zeros, the first element of the vector $\mathbf{\hat{\xi}}_{t+1|t}$ will always equal zero.

Generally, the forecast is then:<br>
<br>
$\hat{y}_{t+1|t} = \mu + \begin{bmatrix} 1 & \theta\end{bmatrix}\begin{bmatrix} \mathbf{\hat{\varepsilon}}_{t+1|t} \\ \mathbf{\hat{\varepsilon}}_{t|t}\end{bmatrix} = \mu + \theta\mathbf{\hat{\varepsilon}}_{t|t}$<br>
<br>
since $\mathbf{\hat{\varepsilon}}_{t+1|t}$ is 0

Now we must update the MSE equation:<br>
<br>
$\mathbf{P}_{t+1|t} = \mathbf{FP}_{t|t}\mathbf{F'} + \mathbf{Q}$<br>
<br>
where $\mathbf{P}_{t|t}$ is updated according to:<br>
<br> 
$\mathbf{P}_{t|t} = \mathbf{P}_{t|t-1} - \mathbf{P}_{t|t-1}\mathbf{H}\left(\mathbf{H'P}_{t|t-1}\mathbf{H} + R\right)^{-1}\mathbf{H'P}_{t|t-1}$<br>
<br>
so in general,<br>
<br>
$\mathbf{P}_{t+1|t} = \begin{bmatrix} 0 & 1 \\ 0 & 0\end{bmatrix}\mathbf{P}_{t|t}\begin{bmatrix} 0 & 0 \\ 1 & 0\end{bmatrix} + \begin{bmatrix} \sigma^2 & 0 \\ 0 & 0\end{bmatrix} = \begin{bmatrix} \sigma^2 & 0 \\ 0 & p_{t+1}\end{bmatrix}$<br>
<br>
where the (2, 2) element of $\mathbf{P}_{t+1|t}$, denoted as $p_{t+1}$, is the same as the (1, 1) element of $\mathbf{P}_{t|t}$<br>
<br>
$\mathbf{P}$ is basically the autocovariance matrix of the forecasted variables. So the variance of $\mathbf{\hat{\varepsilon}}_{t+1|t}$ is always $\sigma^2$ and the forecast error ($\mathbf{\varepsilon}_{t+1} - \mathbf{\hat{\varepsilon}}_{t+1|t}$) is uncorrelated to $(\mathbf{\varepsilon}_{t} - \mathbf{\hat{\varepsilon}}_{t|t})$.

The MSE of the forecast of $y_{t+1}$ is given by:<br>
<br>
$E(y_{t+1} - \hat{y}_{t+1|t})^2 = \mathbf{H'P}_{t+1|t}\mathbf{H} + \mathbf{R}$<br>
<br>
$= \begin{bmatrix} 1 & \theta \end{bmatrix}\begin{bmatrix} \sigma^2 & 0 \\ 0 & p_{t+1}\end{bmatrix}\begin{bmatrix} 1 \\ \theta \end{bmatrix} + 0 = \sigma^2 + \theta^2p_{t+1}$

## Applying the Kalman filter to an AR(1) process
Hamilton recommends the following state space representation for an AR(p) process (see p. 374):<br>

**State Equation** (r = p): <br>
<br>
$\qquad \begin{bmatrix} y_{t+1} - \mu \\ y_t - \mu \\ \vdots \\ y_{t - p +2} - \mu \end{bmatrix} = \begin{bmatrix}\phi_1 & \phi_2 & \cdots & \phi_{p-1} & \phi_p \\ 1 & 0 & \cdots & 0 & 0 \\ 0 & 1 & \cdots & 0 & 0 \\ \vdots & \vdots & \cdots & \vdots & \vdots \\ 0 & 0 & \cdots & 1 & 0 \end{bmatrix}\begin{bmatrix} y_t - \mu \\ y_{t-1} - \mu \\ \vdots \\ y_{t - p + 1} - \mu \end{bmatrix} + \begin{bmatrix}\varepsilon_{t+1} \\ 0 \\ \vdots \\ 0 \end{bmatrix}$ <br><br>
**Observation Equation** (n = 1): <br>
<br>
$~~~~~~~~~~~~~~~y_t = \mu + \begin{bmatrix} 1 & 0 & 0 & \cdots & 0 \end{bmatrix} \begin{bmatrix} y_t - \mu \\ y_{t-1} - \mu \\ \vdots \\ y_{t - p + 1} - \mu \end{bmatrix}$ <br>
<br>
** Terms **: <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{\xi}_t = \begin{bmatrix} y_t - \mu \\ y_{t-1} - \mu \\ \vdots \\ y_{t - p + 1} - \mu \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{F} = \begin{bmatrix}\phi_1 & \phi_2 & \cdots & \phi_{p-1} & \phi_p \\ 1 & 0 & \cdots & 0 & 0 \\ 0 & 1 & \cdots & 0 & 0 \\ \vdots & \vdots & \cdots & \vdots & \vdots \\ 0 & 0 & \cdots & 1 & 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{v}_{t + 1} = \begin{bmatrix}\varepsilon_{t+1} \\ 0 \\ \vdots \\ 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{Q} = E(\mathbf{v}_t\mathbf{v}_t') = \begin{bmatrix} \sigma^2 & 0 & \cdots & 0 \\ 0 & 0 & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{w}_t = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{R} = E(\mathbf{w}_t\mathbf{w}_t') = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{y}_t = y_t$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{A'} = \mu$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{x}_t = 1$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{H'} = \begin{bmatrix} 1 & 0 & 0 & \cdots & 0 \end{bmatrix} \mathbf{\xi}_t$
<br><br>
Therefore, the state-space representation for an AR(1) process is simply:<br><br>
**State Equation** (r = 1): <br><br>
$\qquad \begin{bmatrix} y_{t+1} - \mu \end{bmatrix} = \begin{bmatrix} \phi_1 \end{bmatrix} \begin{bmatrix} y_t - \mu \end{bmatrix} + \begin{bmatrix} \varepsilon_{t+1} \end{bmatrix}$<br><br>
**Observation Equation** (n = 1): <br>
<br>
$~~~~~~~~~~~~~~~y_t = \mu + \begin{bmatrix} 1 \end{bmatrix} \begin{bmatrix} y_t - \mu \end{bmatrix}$ <br><br>
** Terms **: <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{\xi}_t = \begin{bmatrix} y_t - \mu \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{F} = \begin{bmatrix} \phi_1 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{v}_{t + 1} = \begin{bmatrix}\varepsilon_{t+1}\end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{Q} = E(\mathbf{v}_t\mathbf{v}_t') = \begin{bmatrix} \sigma^2 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{w}_t = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{R} = E(\mathbf{w}_t\mathbf{w}_t') = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{y}_t = y_t$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{A'} = \mu$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{x}_t = 1$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{H'} = \begin{bmatrix} 1 & 0 & 0 & \cdots & 0 \end{bmatrix} \mathbf{\xi}_t$
<br><br>

## Applying the Kalman filter to an ARMA(p, q) process
Now, let's work with an ARMA(p, q) process. <br>
Hamilton recommends the following state space representation (see p. 375-376 and [this](https://stats.stackexchange.com/questions/260542/state-space-representation-of-armap-q-from-hamilton) if confused): 

**State Equation** (r = max{p, q + 1}): <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{\xi}_{t + 1} = \begin{bmatrix}\phi_1 & \phi_2 & \cdots & \phi_{r-1} & \phi_r \\ 1 & 0 & \cdots & 0 & 0 \\ 0 & 1 & \cdots & 0 & 0 \\ \vdots & \vdots & \cdots & \vdots & \vdots \\ 0 & 0 & \cdots & 1 & 0 \end{bmatrix}\mathbf{\xi}_t + \begin{bmatrix}\varepsilon_{t+1} \\ 0 \\ \vdots \\ 0 \end{bmatrix}$ <br>
<br>
**Observation Equation** (n = 1): <br>
<br>
$~~~~~~~~~~~~~~~y_t = \mu + \begin{bmatrix} 1 & \theta_1 & \theta_2 & \cdots & \theta_{r-1} \end{bmatrix} \mathbf{\xi}_t$ <br>
<br>
** Terms **: <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{\xi}_t = \phi_1\mathbf{\xi}_{t-1} + \cdots + \phi_r\mathbf{\xi}_{t-r} + \varepsilon_t $<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{F} = \begin{bmatrix}\phi_1 & \phi_2 & \cdots & \phi_{r-1} & \phi_r \\ 1 & 0 & \cdots & 0 & 0 \\ 0 & 1 & \cdots & 0 & 0 \\ \vdots & \vdots & \cdots & \vdots & \vdots \\ 0 & 0 & \cdots & 1 & 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{v}_{t + 1} = \begin{bmatrix}\varepsilon_{t+1} \\ 0 \\ \vdots \\ 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{Q} = E(\mathbf{v}_t\mathbf{v}_t') = \begin{bmatrix} \sigma^2 & 0 & \cdots & 0 \\ 0 & 0 & \cdots & 0 \\ \vdots & \vdots & \cdots & \vdots \\ 0 & 0 & \cdots & 0 \end{bmatrix}$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{w}_t = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{R} = E(\mathbf{w}_t\mathbf{w}_t') = 0$ <br>
<br>
$~~~~~~~~~~~~~~~\mathbf{y}_t = y_t$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{A'} = \mu$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{x}_t = 1$<br>
<br>
$~~~~~~~~~~~~~~~\mathbf{H'} = \begin{bmatrix} 1 & \theta_1 & \theta_2 & \cdots & \theta_{r-1} \end{bmatrix} \mathbf{\xi}_t$
<br>

Let's specify an ARMA(2, 2) process with the following parameters: <br>
$~~~~~~~~~~~~~~\phi_1 = 0.5, \phi_2 = 0.3, \theta_1 = 0.2, \theta_2 = 0.5$ <br>
In equation form that's: <br>
$~~~~~~~~~~~~~~Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \varepsilon_t + \theta_1\varepsilon_{t-1} + \theta_2\varepsilon_{t-2}$ <br>
The first step is to initialize $\mathbf{\hat{\xi}}_{1|0}$ and $\mathbf{P}_{1|0}$<br>
On page 378, Hamilton shows that, provided the eigenvalues of F are inside the unit circle, the Kalman filter iterations can be started with: <br>
$~~~~~~~~~~~~~\mathbf{\hat{\xi}}_{1|0} = \mathbf{0}$<br> 
$~~~~~~~~~~~~~vec(\mathbf{P}_{1|0}) = [\mathbf{I}_{r^2} - (F \otimes F)]^{-1}\cdot vec(\mathbf{Q})$ <br>
where vec represents the <a href="https://en.wikipedia.org/wiki/Vectorization_(mathematics)">vectorization</a> of the given matrix and $\otimes$ represents the [Kronecker product](https://en.wikipedia.org/wiki/Kronecker_product).

```python
import numpy as np

def vec(A):
    return A.transpose().flatten()

nsteps = 1000 # number of steps to simulate

c = 0  # constant in ARMA(2,2) equation

phis = [0.3, -0.2]
thetas = [0.3, 0.1]

r = max(len(phis), len(thetas) + 1)

# initialize all the terms
F = np.zeros([r, r])
F[0, :len(phis)] = phis
for i in range(1, r):
    F[i, i - 1] = 1

sigma = 1  # standard deviation of Gaussian white noise (innovation process)
Q = np.zeros([r, r])
Q[0, 0] = sigma ** 2

wt = np.zeros([1, r])
R = 0
xt = 1
H = np.zeros([r])
H[0] = 1
H[1:] = thetas[:r]

# initialize xi and P based on Hamilton recommendations
xi = np.zeros([nsteps - 1, r])  # xi_1|0 (r x 1)

# Good place to double check math
kron = np.kron(F, F)  # Kronecker Product
I = np.identity(r**2)

vecP = np.linalg.inv(I - kron) @ vec(Q)

P = np.zeros([nsteps - 1, r, r])
P[0, ...] = vecP.reshape(r, r).transpose()  # P_1|0
```

```python
# Check that the eigenvalues of F are inside the unit circle
w, v = np.linalg.eig(F)
print('Eigenvalues: %s' % w)
print('All values lie within unit circle? %s' %(np.max(np.absolute(w)) < 1))
```

We must also generate a realization of this ARMA process which we will use with the Kalman filter. Since this process relies on previous values in the time series, we must make some up. A good choice is to use the unconditional mean (see derivation [here](https://www.kevinsheppard.com/images/c/cf/Chapter4.pdf)):<br>
 $~~~~~~~~~~~E[\mathbf{y}_t] = \dfrac{c}{1 - \sum_{p=1}^{p} \phi_p}$

```python
A = c / (1 - sum(phis))  # unconditional mean
yt = np.zeros([nsteps + len(phis)])
et = np.zeros_like(yt)  # keep a record of the previous white noise values
yt[:2] = A

for i in range(len(phis), yt.size):
    et[i] = np.random.normal()
    yt[i] = c + phis[0]*yt[i-1] + phis[1]*yt[i-2] + et[i] + thetas[0]*et[i-1] + thetas[1]*et[i-2]

yt = yt[len(phis):] # cut out initialization steps

# plot the trajectory
import matplotlib.pyplot as plt

plt.plot(np.arange(nsteps)[:100], yt[:100])
plt.show()
```

Now we can begin forecasting. We will start by calculating $\mathbf{\hat{\xi}}_{2|1}$ and $\mathbf{P}_{2|1}$. <br>
First, forecast the value of $\mathbf{\hat{y}}_t$:<br>
$~~~~~~~~~~~\mathbf{\hat{y}}_{t|t-1} = \mathbf{A}'\mathbf{x}_t + \mathbf{H}'\mathbf{\hat{\xi}}_{t|t-1}$<br>
Before finding $\mathbf{\hat{\xi}}_{2|1}$, we must update the inference about the current value of $\mathbf{\xi}_1$ based on the actual observed value of $\mathbf{y}_1$. The following generalized equation can achieve this:<br>
$~~~~~~~~~~~\mathbf{\hat{\xi}}_{t|t} = \mathbf{\hat{\xi}}_{t|t - 1} + \mathbf{P}_{t|t-1}\mathbf{H}(\mathbf{H}'\mathbf{P}_{t|t-1}\mathbf{H} + \mathbf{R})^{-1}(\mathbf{y}_t - \mathbf{A}'\mathbf{x}_t - \mathbf{H}'\mathbf{\hat{\xi}}_{t|t-1})$<br>
Note that the above equation implicitly uses the equation previous to it. 

```python
xi_t = np.zeros_like(xi)  # will hold updated values of xi
yt_forecast = np.zeros([nsteps - 1])

yt_forecast[0] = A*xt + H.transpose() @ xi[0, :]  # forecast value of yt

PH = P[0, ...] @ H  # 3 x 1 column vector
HPHR = (H.transpose() @ P[0, ...] @ H + R)**-1
err = yt[1] - yt_forecast[0]

xi_t[0, :] = xi[0, :] + PH * HPHR * err

print("Forecasted value of yt: %.3f" % yt_forecast[0])
print(xi_t[0, :])
```

The MSE associated with this updated projection, $\mathbf{P}_{t|t}$, is: <br>
$~~~~~~~~~~~\mathbf{P}_{t|t} = \mathbf{P}_{t|t-1} - \mathbf{P}_{t|t-1}\mathbf{H}(\mathbf{H'}\mathbf{P}_{t|t-1}\mathbf{H} + \mathbf{R})^{-1}\mathbf{H'}\mathbf{P}_{t|t-1}$

```python
P_t = np.zeros_like(P)

# using values from above
P_t[0, ...] = P[0, ...] - PH * HPHR * (H.transpose() @ P[0, ...])
```

Now, forecast the value of $\mathbf{\hat{\xi}}_{t+1|t}$ using the following equation:<br>
$~~~~~~~~~~~\mathbf{\hat{\xi}}_{t+1|t} = \mathbf{F}\mathbf{\hat{\xi}}_{t|t}$

```python
xi[1, :] = F @ xi_t[0, ...]
```

And forecast the value of $\mathbf{P}_{t+1|t}$:<br>
$~~~~~~~~~~~~\mathbf{P}_{t+1|t} = \mathbf{F}\mathbf{P}_{t|t}\mathbf{F'} + \mathbf{Q}$

```python
P[1, ...] = F @ P_t[0, ...] @ F.transpose() + Q
```

Now, recursively continue the Kalman filter

```python
for s in range(1, nsteps - 1):

    yt_forecast[s] = A*xt + H.transpose() @ xi[s, :]  # forecast value of yt

    PH = P[s, ...] @ H  # 3 x 1 column vector
    HPHR = (H.transpose() @ P[s, ...] @ H + R)**-1   # scalar
    err = yt[s + 1] - yt_forecast[s] # scalar
    
    xi_t[s, :] = xi[s, :] + PH * HPHR * err
    
    P_t[s, ...] = P[s, ...] - (PH * HPHR * (H.transpose() @ P[s, ...]))
    
    if s < (nsteps - 2):
        xi[s + 1, :] = F @ xi_t[s, :]
        P[s + 1, ...] = F @ P_t[s, ...] @ F.transpose() + Q
```

```python
# Plot the forecast versus the actual data

fig, ax = plt.subplots(figsize=(15, 5))

ax.plot(np.arange(nsteps - 1), yt[1:], label='Observed Data', linewidth=2, alpha=0.5)
ax.plot(np.arange(nsteps - 1), yt_forecast, '--', label='Forecasted Data', linewidth=1)

# ax.set_ylim(-10, 35)
ax.set_xlim(0, 200)  # to see the data better
plt.tick_params(labelsize=14)
ax.set_ylabel('yt', fontsize=14)
ax.set_xlabel('t', fontsize=14)
plt.legend(fontsize=14)
plt.show() 
```

# MLE Parameter Estimation
Next, we can predict maximum likelihood estimates of the parameters that make up $\mathbf{F}, \mathbf{Q}, \mathbf{A}, \mathbf{H}$ and $\mathbf{R}$. To do so, we must maximize the likelihood function (p. 385-386): <br>
$~~~~~~~~~~~~~~\sum_{t=1}^{T} log~f_{\mathbf{Y}_t|\mathbf{X}_t, \mathscr{Y}_{t-1}}(\mathbf{y}_t|\mathbf{x}_t, \mathscr{Y}_{t-1})$<br>
where<br>
$~~~~~~~~~~~~~~f_{\mathbf{Y}_t|\mathbf{X}_t, \mathscr{Y}_{t-1}} = (2\pi)^{-n/2}\mid\mathbf{H'}\mathbf{P}_{t|t-1}\mathbf{H} + \mathbf{R}\mid^{-1/2}~\times~e^{-0.5(\mathbf{y}_t - \mathbf{A'}\mathbf{x}_t - \mathbf{H'}\mathbf{\hat{\xi}}_{t|t-1})'(\mathbf{H'}\mathbf{P}_{t|t-1}\mathbf{H} + R)^{-1}(\mathbf{y}_t - \mathbf{A'}\mathbf{x}_t - \mathbf{H'}\mathbf{\hat{\xi}}_{t|t-1})}$<br>
and $n$ is the dimension of $y_t$ (n = 1 in this case)

```python
# In order to maximize the log-likelihood function above, we must contstruct it so it 
# is compatible with scipy minimize. The following function is made up primarily of code
# already written above, with the addition of a likelihood function.

def kalman_log_likelihood(params, yt, p, q, plot=False, n=1):
    """ Calculate log likelihood function of kalman filter applied to timeseries with parameters
    in params. *Note that this is specifically for an ARMA process*
    
    :param params: list of parameters: [phis, thetas, sigma, c, ]
    :param yt: observed time series
    :param p: number of autoregressive terms of yt (i.e y_t-1, t_t-2 etc.)
    :param q: number of moving average terms
    """
    
    # restructure params into F, Q, A, H and R as defined in Hamilton
    phis = params[:p]
    thetas = params[p:(p + q)]
    r = max(len(phis), len(thetas) + 1)

    # initialize all the terms
    F = np.zeros([r, r])
    F[0, :len(phis)] = phis
    for i in range(1, r):
        F[i, i - 1] = 1

    # standard deviation of Gaussian white noise (innovation process)
    #sigma = params[p + q]
    sigma = 1
    Q = np.zeros([r, r])
    Q[0, 0] = sigma ** 2
    
    A = 0

    R = 0  # For general ARMA, in Hamilton's suggested formulation, R = 0

    H = np.zeros([r])
    H[0] = 1
    H[1:] = thetas[:r] 

    if len(yt.shape) == 1:
        n = 1
    else:
        n = yt.shape[1]  # double check once using multi dimensions

    nsteps = yt.size

    # initialize xi based on Hamilton recommendations
    xi = np.zeros([nsteps - 1, r])  # xi_1|0 (r x 1)

    # initialize xi based on Hamilton recommendations
    P = np.zeros([nsteps - 1, r, r])
    kron = np.kron(F, F)  # Kronecker Product
    I = np.identity(r**2)
    vecP = np.linalg.inv(I - kron) @ vec(Q)
    P[0, ...] = vecP.reshape(r, r).transpose()  # P_1|0
    
    xt = 1
    
    yt_forecast = np.zeros([nsteps - 1])
    likelihood = np.zeros([nsteps - 1])
    for s in range(nsteps - 1):

        yt_forecast[s] = A*xt + H.transpose() @ xi[s, :]  # forecast value of yt

        PH = P[s, ...] @ H  # 3 x 1 column vector
        HPHR = (H.transpose() @ P[s, ...] @ H + R)**-1   # scalar
        err = yt[s + 1] - yt_forecast[s] # scalar

        xi_t[s, :] = xi[s, :] + PH * HPHR * err

        P_t[s, ...] = P[s, ...] - (PH * HPHR * (H.transpose() @ P[s, ...]))

        if s < (nsteps - 2):
            xi[s + 1, :] = F @ xi_t[s, :]
            P[s + 1, ...] = F @ P_t[s, ...] @ F.transpose() + Q
        
        # this will need to be updated once multidimensional
        likelihood[s] = np.log((2 * np.pi)**(-n/2)*np.abs(HPHR**-1)**(-0.5)) - 0.5*err*HPHR*err
    
    if plot:
        
        fig, ax = plt.subplots(figsize=(15, 5))

        ax.plot(np.arange(nsteps - 1), yt[1:], label='Observed Data', linewidth=2, alpha=0.5)
        ax.plot(np.arange(nsteps - 1), yt_forecast, '--', label='Forecasted Data', linewidth=1)

        # ax.set_ylim(-10, 35)
        ax.set_xlim(0, 200)  # to see the data better
        plt.tick_params(labelsize=14)
        ax.set_ylabel('yt', fontsize=14)
        ax.set_xlabel('t', fontsize=14)
        plt.legend(fontsize=14)
        plt.show() 

    return -likelihood.sum()  # return negative value since we will use scipy minimize

def eigenvalue_constraint(params):
    """ Ensure eigenvalues are all within unit circle
    """
    
    phis = params[:2]  # there is an args parameter to the constraints option in the minimizer
    # to make this for general values of p and q
    thetas = params[2:]
    r = max(len(phis), len(thetas) + 1)

    # initialize all the terms
    F = np.zeros([r, r])
    F[0, :len(phis)] = phis
    for i in range(1, r):
        F[i, i - 1] = 1

    w, v = np.linalg.eig(F)

    m = int(np.max(np.absolute(w)) < 1)

    return -1 + m

from scipy.optimize import minimize
guess = [0.3, 0.2, 0.1, 0.1]

my_constraints = ({"type": 'eq', "fun": eigenvalue_constraint})

bounds = [(-.99, .99), (-.99, .99), (-.99, .99), (-.99, .99)]
res = minimize(kalman_log_likelihood, guess, args=(yt, 2, 2), bounds=bounds, method = 'TNC')#, constraints=my_constraints)

kalman_log_likelihood(res.x, yt, 2, 2, plot=True)

print("Maximum Likelihood Parameter Estimates:\n")
print("phi_1 = %.2f" %res.x[0], ", actual phi_1 = %.2f" % phis[0])
print("phi_2 = %.2f" %res.x[1], ", actual phi_2 = %.2f" % phis[1])
print("theta_1 = %.2f" %res.x[2], ", actual theta_1 = %.2f" % thetas[0])
print("theta_2 = %.2f" %res.x[3], ", actual theta_2 = %.2f" % thetas[1])
```

# Smoothing
The next subject to understand is Smoothing. See section of 13.6 of Hamilton for full derivations. I will just state the main equations.<br>
<br>
Instead of updating our estimate sequentially based on previous values of the data, we now use knowledge of the entire time series.<br>
<br>
The smoothed estimate of the final time in our sample, $\mathbf{\hat{\xi}}_{T|T}$, is just the last entry in the timeseries $\{\mathbf{\hat{\xi}}_{t|t} \}_{t=1}^T$.

```python
xi_smoothed = np.zeros([xi.shape[0] - 1, r])
xi_smoothed[-1, :] = xi_t[-1, :]
```

Now, we define a paramater $\mathbf{J}_t$:<br>
$~~~~~~~~~~~~\mathbf{J}_t \equiv \mathbf{P}_{t|t}\mathbf{F'}\mathbf{P}_{t+1|t}^{-1}$<br>
and use this definition to generate $\{\mathbf{J}_t\}_{t=1}^{T-1}$.

```python
J = np.zeros([xi.shape[0] - 1, r, r])  # a series of 3x3 matrices

for j in range(nsteps - 2):
    J[j, :] = P_t[j, ...] @ F.transpose() @ np.linalg.inv(P[j + 1, ...])
```

In general we can calculate the value of $\mathbf{\hat{\xi}}$ based on all timeseries values with the following formula: <br>
$~~~~~~~~~~~\mathbf{\hat{\xi}}_{t|T} = \mathbf{\hat{\xi}}_{t|t} + \mathbf{J}_t(\mathbf{\hat{\xi}}_{t + 1|T} - \mathbf{\hat{\xi}}_{t + 1|t})$<br>
We start by using the above equation for $t = T - 1$:<br>
$~~~~~~~~~~~\mathbf{\hat{\xi}}_{T-1|T} = \mathbf{\hat{\xi}}_{T-1|T-1} + \mathbf{J}_{T-1}(\mathbf{\hat{\xi}}_{T|T} - \mathbf{\hat{\xi}}_{T|T-1})$<br>
And now that $\mathbf{\hat{\xi}}_{T-1|T}$ has been calculated, we can solve for $t = T - 2$: <br>
$~~~~~~~~~~~\mathbf{\hat{\xi}}_{T-2|T} = \mathbf{\hat{\xi}}_{T-2|T-2} + \mathbf{J}_{T-2}(\mathbf{\hat{\xi}}_{T-1|T} - \mathbf{\hat{\xi}}_{T-1|T-2})$<br>
And now the recursion continues backwards to generate the smoothed sequence $\{\mathbf{\hat{\xi}}_{t|T}\}_{t=1}^T$

```python
for i in range(nsteps - 4, -1, -1):  # work backwards down to zero starting at second to last entry of xi_smoothed
    xi_smoothed[i, :] = xi_t[i + 1, :] + J[i + 1, ...] @ (xi_t[i + 2, :] - xi[i + 2, :])
```

Finally, we must calculate the mean squared error associated with this projection:<br>
$~~~~~~~~~~~\mathbf{P}_{t|T} = \mathbf{P}_{t|t} + \mathbf{J}_t(\mathbf{P}_{t + 1|T} - \mathbf{P}_{t + 1|t})\mathbf{J'}_t$<br>
Again, this sequence is generated by moving through the sample backwards starting with $t = T - 1$.

```python
P_smoothed = np.zeros([nsteps - 2, r, r])
P_smoothed[-1, ...] = P_t[-1, ...]

for i in range(nsteps - 4, -1, -1):
    P_smoothed[i, ...] = P_t[i + 1, :] + J[i + 1, ...] @ (P_t[i + 2, ...] - P[i + 2, ...]) @ J[i + 1, ...].transpose()
```

## AR(1) Process
Let's design a filter for an AR(1) process. The recursive relationship looks like:<br>
$\qquad z_t = \phi z_{t-1} + \varepsilon_t$<br>
<br>Let's first create a signal based on this relationship, so that we can process it with a Kalman filter.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_AR1(phi, sigma, dt, N):
    """ Generate an autoregressive time series of order 1
    """
    
    z = [sigma*np.random.randn()]  # initialize with a random number chosen from a mean centered at zero
    for i in range(1, N):
        z.append(phi * z[i - 1] + sigma*np.random.randn())
        
    time = dt * np.arange(N)
    
    return z, time

N = 100  # number of steps to take
dt = 1
phi = 0.3
sigma = 1  # standard deviation of white noise

z, time = generate_AR1(phi, sigma, dt, N)

# Plot our time series
plt.plot(time, z, linewidth=2)
plt.ylabel('$z$', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.tick_params(labelsize=14)
plt.show()
```

Now we need to think about how to set up the Kalman filter. Following the notation given by Roger Labbe, the state vector can be written as:<br>
$\qquad \mathbf{x} = \begin{bmatrix} z_t \\ z_{t-1} \end{bmatrix}$<br>
The process model $\mathbf{F}$ propagates the state and can be written like so:<br>
$\qquad \mathbf{F} = \begin{bmatrix} 0 & \phi \\ 0 & 1 \end{bmatrix}$<br>
since<br>
$\qquad \mathbf{\bar{x}} = \mathbf{F}\mathbf{x} = \begin{bmatrix} 0 & \phi \\ 0 & 1 \end{bmatrix}\begin{bmatrix} z_t \\ z_{t-1} \end{bmatrix}$  ($\mathbf{\bar{x}}$ denotes a prediction)<br>
which yields the system of equations:<br>
$\qquad z_t = \phi z_{t-1}$<br>
$\qquad z_{t-1} = z_{t-1}$<br>
Our measurement, $z_t$, can be represented in matrix form as:<br>
$\qquad \begin{bmatrix} z_t \end{bmatrix}$<br>
since we are only getting one measurement, which is the value of z.<br>
<br>
First, we will initialize these values with the KalmanFilter class of filterpy:

```python
from filterpy.kalman import KalmanFilter

xinit = [0, 0]  # initial guess at y_t and y_t-1 (using the unconditional expectation here)

kf = KalmanFilter(dim_x=len(xinit), dim_z=1)
kf.x = np.array([xinit[0], xinit[1]])
kf.F = np.array([[0, phi], [0, 1]])
```

We will also need to define the process noise, $\mathbf{Q}$, the measurement function, $\mathbf{H}$, and the measurement noise $\mathbf{R}$<br><br>
A useful approximation for $\mathbf{Q}$ is <br><br>
$\qquad \mathbf{Q} = \begin{bmatrix} 0 & 0 \\ 0 & \sigma^2 \end{bmatrix}$<br><br>
However, I'll use `filterpy`'s implentation of *better* guess (see code below).<br><br>
The measurement function defines how we go from the state variables to the measurements and is used to calculate the residual, $\mathbf{y}$ (the difference between the prediction and actual measurement), with the equation $\mathbf{y} = \mathbf{z} - \mathbf{H}\mathbf{x}$. Since $\mathbf{z}$ is $(1 \times 1)$ and our $\mathbf{x}$ is $(2\times 1)$, we can determine the size of $\mathbf{H}$ by reasoning that<br><br>
$\qquad (1\times 1) = (a\times b)(2\times 1) = (1\times 2)(2\times 1)$<br><br>
so $\mathbf{H}$ is $(1\times 2)$. In this case,<br><br>
$\qquad \mathbf{H} = \begin{bmatrix} 1 & 0 \end{bmatrix}$

In this way, our residual is the difference between $z_t$ and the predicted value of $z_t$ since<br><br>
$\qquad \mathbf{z} - \mathbf{H}\mathbf{\bar{x}} = \begin{bmatrix} 1 & 0 \end{bmatrix}\begin{bmatrix} \bar{z}_t \\ \bar{z}_{t-1} \end{bmatrix} = z_t - \bar{z}_t$

We will assume no measurement noise:<br>

$\qquad \mathbf{R} = \mathbf{0}$

```python
from filterpy.common import Q_discrete_white_noise

kf.Q = Q_discrete_white_noise(dim=2, dt=dt, var=sigma)
kf.R *= 0
kf.H = np.array([[1, 0]])
```

Okay, now we can filter

```python
xs, cov = [], []

for i in z:
    kf.predict()
    kf.update(i)
    xs.append(kf.x)
    cov.append(kf.P)
    
xs, cov = np.array(xs), np.array(cov)

plt.plot(time, z, color='blue', label='Measurements', linewidth=2)
plt.plot(time, xs[:, 0], color='orange', label='Filter', linewidth=2)
plt.legend(fontsize=14)
plt.show()
```

`filterpy` calculates the log-likelihood of a fit with<br>
$\qquad \mathcal{L} = \frac{1}{\sqrt{2\pi \mathbf S}}\exp [-\frac{1}{2}\mathbf y^\mathsf T\mathbf S^{-1}\mathbf y]$

Let's find the optimal $\phi$ parameter using this likelihood function. First, we will write the kalman filter into a function which accepts $\phi$ as an argument:

```python
from filterpy.kalman import KalmanFilter
from filterpy.common import Q_discrete_white_noise

def AR_filter(zs, xinit, phi):
    """ Filter measurements z as if it were an AR(1) process defined by phi
    
    :param z: measurements
    :param xinit: initial guess at x_t and x_t-1
    :param phi: coefficient of autoregressive term in z_t = phi*z_t-1 + eps
    """

    kf = KalmanFilter(dim_x=len(xinit), dim_z=1)
    kf.x = np.array([xinit[0], xinit[1]])
    kf.F = np.array([[0, phi], [0, 1]])
    kf.Q = Q_discrete_white_noise(dim=2, dt=1, var=sigma**2)
    kf.R *= 0
    kf.H = np.array([[1, 0]])
    
    xs, cov = [], []
    
    for z in zs:
        kf.predict()
        kf.update(z)
        xs.append(kf.x)
        cov.append(kf.P)
    
    xs, cov = np.array(xs), np.array(cov)
    
    return kf, xs, cov

# parameters to generate AR(1) timeseries
N = 50  # number of steps to take
dt = 1
phi = -0.2
sigma = 0.10  # standard deviation of white noise

# parameter for AR_filter
xinit = np.array([0, 0])
z, time = generate_AR1(phi, 0.1, dt, N)
phis = np.linspace(-0.8, 0.8, 17)

f, xs, cov = AR_filter(z, xinit, phi)

plt.plot(time, z)
plt.plot(time, xs[:, 0])

likelihood = []
for i in phis:
    
    f, xs, cov = AR_filter(z, xinit, i)
    
    likelihood.append(f.log_likelihood)

plt.figure()
plt.plot([phi, phi], [min(likelihood), max(likelihood)], '--', color='black')
plt.plot(phis, likelihood)
plt.show()

# from statsmodels.tsa.arima_model import ARIMA

# model = ARIMA(z, order=(1, 0, 0))
# model_fit = model.fit()
# model.loglike_kalman(z) 
```
